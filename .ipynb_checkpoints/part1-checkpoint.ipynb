```python
import plotly as py
import pandas as pd
import comet_ml
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats


#comet config
comet_ml.init()

#importing dataset
plData = pd.read_csv("EPL_20_21.csv")
print(plData.head())
```

```python
redyello = plData[['Club','Red_Cards','Yellow_Cards']]
df_1 = redyello.groupby('Club', as_index=False).agg({"Red_Cards": "sum", "Yellow_Cards": "sum"})
countries = df_1["Club"]
#red_cards = df_1['Red_Cards'].tolist()
red_cards = df_1['Red_Cards']
yellow_cards = df_1['Yellow_Cards']

width = 0.75

 

fig, ax = plt.subplots(figsize=(16, 10))

ax.bar(countries, yellow_cards, width, label='Yellow Cards', color='gold')
ax.bar(countries, red_cards, width, bottom=yellow_cards, label='Red Cards', color='orangered')

ax.set_ylabel('Amount of Cards', fontsize=14)
ax.set_title('Number of Yellow and Read Cards in Premier League (2020/21)', loc='left', fontsize=18, fontweight ='bold')
plt.xticks(countries, rotation=90, fontsize=12)

 

for index, data in enumerate(red_cards):
    plt.text(x=index , y=data + yellow_cards[index] + 1 , s=f"{data}" , fontdict=dict(fontsize=14), horizontalalignment='center')

 

for index, data in enumerate(yellow_cards):
    plt.text(x=index , y=20 , s=f"{data}" , fontdict=dict(fontsize=14), horizontalalignment='center')
    
ax.legend(fontsize=12, frameon=False)

fig.text(0.9, -0.08, 'mariuszborycki.com', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)

plt.show()
```

```python
#utility functions
#return unique values of a list
def uniquef(list1):
    npArray1 = np.array(list1)
    uniqueNpArray = np.unique(npArray1)
    return uniqueNpArray.tolist()
```

```python
#Now lets define a function that calculates the point based on fantasy premier league system
#source : https://qr.ae/pvCL53
#calculating rating function (to be edited)
def calculate_points(data):
    pts = 0
    avgMins = data["Mins"]//data["Matches"]
    pts += data["Assists"]*3
    if avgMins == 60:    
        pts += data["Matches"]
    elif avgMins > 60:
        pts += data["Matches"]*3
    pts -= data["Yellow_Cards"]
    pts -= data["Red_Cards"]*3
    if data["Position"] == "FW":
        pts += data["Goals"]*4
    elif data["Position"] == "MF":
        pts += data["Goals"]*3
    
        
```
